```python
import os
from dotenv import load_dotenv
load_dotenv()
OPEN_AI_KEY = os.getenv('OPEN_AI_KEY')
```

```python
from openai import OpenAI
import pprint

client = OpenAI(api_key=OPEN_AI_KEY)
response = client.chat.completions.create(
  model="gpt-4",
  messages=[
    {
      "role": "system",
      "content": """Given the following SQL tables, your job is to write a SQL query given a user’s request. The query MUST NOT make changes to the database.IT MUST ONLY BE A SELECT STATEMENT. MAKE SURE YOU ONLY RESPOND WITH THE SQL QUERY AND NOTHING ELSE.AND MAKE YOUR ANSWER SINGLE LINED.\n    \n    
      
      CREATE TABLE youtube.cities_chart(
            id integer,
            city_name varchar,
            date DATE,
            views integer
      );\n    \n 
      CREATE TABLE youtube.cities_aggregate(
            id integer,
            city_name varchar,
            geography varchar,
            watch_time_in_hours float,
            views integer,
            average_view_duration interval
      );\n    \n 
      CREATE TABLE youtube.device_type_chart(
            id integer,
            name varchar,
            date DATE,
            views integer
      ); \n    \n 
      CREATE TABLE youtube.device_type_aggregate(
        id integer,
        name varchar,
        watch_time_in_hours float,
        views integer,
        average_view_duration interval
      );\n    \n 
        """
    },
    {
      "role": "user",
      "content": "Give me the city and country where we were watched from the most."
    }
  ],
  temperature=0.7,
  max_tokens=64,
  top_p=1
)
pprint.pprint(response.choices[0].message.content)
```
